```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx

import tensorflow as tf
import tensorflow_hub as hub
```

```python
df = pd.read_csv("wine-reviews.csv", usecols = ['country','points','price'])
```

```python
plt.hist(df.points,bins=20)
plt.title("Points Histrogram")
plt.ylabel("N")
plt.xlabel("Points")
plt.show()
```

```python
# create a graph using Source and Target for connections.
G = nx.from_pandas_edgelist(df,'country','points','price')

# Define the positions of the nodes
pos = nx.spring_layout(G)

# Draw the graph with customized parameters
nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=500,  # Adjust the node size
    node_color='lightblue',  # Set the node color
    edge_color='gray',  # Set the edge color
    width=1.5,  # Adjust the edge width
    font_size=10,  # Adjust the font size of labels
    alpha=0.8,  # Set the transparency level
    arrowsize=10,  # Adjust the size of arrowheads (if applicable)
)

# Remove the axis labels
plt.axis('off')

# Show the plot
plt.show()
```
